```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("sample_data/social_ads_dataset.xlsx")
```

```python
df
```

```python
target = 'sales'
inputs = ['google', 'facebook','instagram','size']

X = df[inputs]
y = df[target]
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(X,y)
```

```python
lm.score(X, y)
```
