```python
# Import library to read CSV file
import pandas as pd

# Import library to load data into BigQuery
import pandas_gbq

# Import Google Authentication Library
from pydrive.auth import GoogleAuth
from google.colab import auth

# Authenticate Current User
auth.authenticate_user()
gauth = GoogleAuth()
```

```python
project_id = "aibe-amirul-march2023"
dataset = "sources"
table_name = "hr_dataset"
```

```python
# Read CSV file with Pandas
df = pd.read_csv("sample_data/HR_dataset.csv")
# Show the top 5 rows
df.head()
```

```python
df.columns
```

```python
# Change column names
df.columns = ['names', 'satisfaction_level', 'last_evaluation', 'number_project',
       'average_monthly_hours', 'exp_in_company', 'work_accident', 'left_company',
       'promotion_last_5years', 'role', 'salary']
```

```python
# Select Columns that we need
df = df[['names', 'satisfaction_level', 'last_evaluation', 'number_project',
       'average_monthly_hours', 'exp_in_company', 'work_accident', 'left_company',
       'promotion_last_5years', 'role', 'salary']]
```

```python
df.head()
```

```python
# upload dataset into dataframe in BigQuery

pandas_gbq.to_gbq(
    df,
    f"{dataset}.{table_name}",
    project_id=project_id,
    if_exists="replace"
)
```
